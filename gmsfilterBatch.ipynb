```python
def computefeat(temp,direcX):
    feature = [[0 for i in range(len(temp))] for j in range(len(temp))] 

    for i in range(0,len(temp)):
        for j in range(0,len(temp)):
            img1 = (np.array(direcX[0][i]))
            img2 = (np.array(direcX[0][j]))

            if(i==j):
                continue

            if(i!=j):
                orb = cv2.ORB_create(10000)
                orb.setFastThreshold(0)
                if cv2.__version__.startswith('3'):
                    matcher = cv2.BFMatcher(cv2.NORM_HAMMING)
                else:
                    matcher = cv2.BFMatcher_create(cv2.NORM_HAMMING)
                gms = GmsMatcher(orb, matcher)

                matches = gms.compute_matches(img1, img2)
                #gms.draw_matches(img1, img2, DrawingType.ONLY_LINES)
                #gms.draw_matches(img1, img2, DrawingType.LINES_AND_POINTS)
                #print(len(matches))           
                feature[i][j]=len(matches)
    return feature
```

```python
data_src= './data_repository/Veri776/'
X = []
y = []
direcX=[]
direcy=[]
F=[]
id=[]
for directory in os.listdir(data_src):
    print(directory)
    temp=[]
    c=0
    try:
        for pic in os.listdir(os.path.join(data_src, directory)):
            #img = imread(os.path.join(self.data_src, directory, pic))
            #img = Image.open(os.path.join(self.data_src, directory, pic))
            #img=img.resize((224,224))
            
            img = cv2.imread(os.path.join(data_src, directory, pic))
            img = cv2.resize(img, (224,224))
            #print(img)

            temp.append(np.squeeze(np.asarray(img)))
            
            X.append(np.squeeze(np.asarray(img)))
            y.append(directory)
            id.append(c)
            c=c+1
        direcX.append(np.array(temp))
        direcy.append(directory)
        F.append(np.squeeze(np.asarray(computefeat(temp,direcX))))
        #print(direcX)
    except Exception as e:
        print('Failed to read images from Directory: ', directory)
        print('Exception Message: ', e)
print('Dataset loaded successfully.')
```

```python

# Setup Network
anchor_input = tf.placeholder(tf.float32, placeholder_shape, name='anchor_input')
positive_input = tf.placeholder(tf.float32, placeholder_shape, name='positive_input')
negative_input = tf.placeholder(tf.float32, placeholder_shape, name='negative_input')

margin = 0.5
anchor_output = model.conv_net2(anchor_input, reuse=False)
positive_output = model.conv_net2(positive_input, reuse=True)
negative_output = model.conv_net2(negative_input, reuse=True)
loss = model.triplet_loss(anchor_output, positive_output, negative_output, margin)

# Setup Optimizer
global_step = tf.Variable(0, trainable=False)

train_step = tf.train.MomentumOptimizer(0.01, 0.99, use_nesterov=True).minimize(loss, global_step=global_step)
```

```python
from preprocessing import PreProcessing
from model import TripletLoss
next_batch = dataset.get_triplets_batch
```

```python
# Start Training
saver = tf.train.Saver()
with tf.device('/gpu:1'):

    with tf.Session(config=tf.ConfigProto(allow_soft_placement=True, log_device_placement=True)) as sess:
        sess.run(tf.global_variables_initializer())

        # Setup Tensorboard
        tf.summary.scalar('step', global_step)
        tf.summary.scalar('loss', loss)
        for var in tf.trainable_variables():
            tf.summary.histogram(var.op.name, var)
        merged = tf.summary.merge_all()
        writer = tf.summary.FileWriter('train.log', sess.graph)

        # Train iter
        for i in range(1):
            batch_anchor, batch_positive, batch_negative = next_batch(2)
           # print(batch_anchor)
            
            _, l, summary_str = sess.run([train_step, loss, merged],
                                         feed_dict={anchor_input: batch_anchor, positive_input: batch_positive, negative_input: batch_negative})

            writer.add_summary(summary_str, i)
            print("\r#%d - Loss" % i, l)

            if (i + 1) % FLAGS.step == 0:
                saver.save(sess, "model_triplet/model128.ckpt")
        saver.save(sess, "model_triplet/model128.ckpt")
print('Training completed successfully.')
```
